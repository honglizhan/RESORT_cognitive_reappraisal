```python
import pandas as pd
human_eval_df = pd.read_json("human_eval_annotations_cleaned.json")
print (len(human_eval_df))
human_eval_df.head(1)
```

```python
cleaned_rows = []

for _, row in human_eval_df.iterrows():
    row_dict = row.to_dict()
    annotation_keys = [key for key in row.keys() if key.startswith('human_eval_annotation_')]

    for key in annotation_keys:
        annotation = row[key]
        if annotation:
            # Merge row_dict and annotation dictionaries
            cleaned_row = {**row_dict, **annotation}  # Combine dictionaries
            cleaned_rows.append(cleaned_row)

cleaned_df = pd.DataFrame(cleaned_rows)
print (len(cleaned_df))
cleaned_df.head(2)
```

## Mean and Std for all score columns

```python
group_columns = ["model", "pipeline", "method"]
score_columns = ["alignment_with_standards", "empathy", "harmfulness", "factuality"]

# Calculate mean and std for each score column
score_stats = cleaned_df.groupby(group_columns)[score_columns].agg(['mean', 'std']).reset_index()

score_stats.columns = ['_'.join(col).strip('_') if isinstance(col, tuple) else col for col in score_stats.columns]
score_stats.to_excel("table3_results.xlsx", index=False)
```
